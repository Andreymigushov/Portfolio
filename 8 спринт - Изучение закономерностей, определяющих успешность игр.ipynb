### Предобработка данных

**2.1.** Заменим названия столбцов (приведем к нижнему регистру).

```python
data.columns = data.columns.str.lower()
```

**2.2.** Проанализируем столбец 'user_score'.

Из полученной информации о последних 15 строках мы видим, что в столбце 'user_score' кроме числовых значений и NaN, иногда встречаются значения 'tbd'. Проверим этот столбец на наличие уникальных значений.

```python
data['user_score'].unique()
```

Убеждаемся, что в столбце 'user_score' встречаются числовые значения, 'NaN' и 'tbd'. Аббревиатура 'tbd' означает рейтинг не определен. Можно просто заменить эти значения на 'NaN'. Так и сделаем. Заменим значения 'tbd' на 'NaN'.

```python
data['user_score'] = data['user_score'].str.replace('tbd','NaN')
```

**2.3.** Поменяем тип данных в столбцах 'year_of_release' и 'user_score'.

Поменяем тип данных для: 'year_of_release' - на Int16, 'user_score' - на float64. Изменения на типы Int16 и float64 вызваны необходимостью удобства работы и экономии памяти.

```python
data['year_of_release'] = data['year_of_release'].astype('Int16')
```

```python
data['user_score'] = data['user_score'].astype('float64')
```

**2.4.** Проанализируем столбец 'year_of_release'.

Проверим столбец 'year_of_release' на наличие уникальных значений.

```python
data['year_of_release'].unique()
```

Убеждаемся, что в столбце 'year_of_release' кроме числовых значений встречаются значения 'NA' (пропущенные значения).

Подсчёт количества пропущенных значений в столбце 'year_of_release'.

```python
data['year_of_release'].isna().sum()
```

В столбце 'year_of_release' количество пропущенных значений составляет около 1,61 %. Доля этих значений крайне мала. Удалим эти строки. (Возможно это произошло из-за того, что просто забыли указать год выпуска игры или дата выпуска игры точно не известна).

```python
data['year_of_release'].isna().sum()
```

**2.5.** Проанализируем столбец 'name'.

Проерим столбец 'name' на наличие дубликатов.

```python
data['name'].duplicated().sum()
```

Возможно одни и те же игры выпускали с обновлениями в разные годы.

```python
data[data[['name', 'platform', 'year']].duplicated()]
```

Удалим эти дубликаты.

```python
data = data.drop_duplicates(subset=['name', 'platform', 'year'], keep='first')
```

Проверим столбец 'name' на наличие уникальных значений.

```python
data['name'].nunique() # проверим количество всех уникальных значений в столбце
```

**2.6.** Проанализируем столбец 'rating'.

 Получим набор всех уникальных значений в столбце 'rating'.

```python
data['rating'].unique()
```

Все эти обозначения (кроме 'NaN') соответствуют определению рейтингов для компьютерных видеоигр и другого развлекательного программного обеспечения в США и Канаде негосударственной организацией Entertainment Software Rating Board (ESRB). Заменим эти обозначения на числовые значения со знаком "+"  (для удобства чтения таблицы):   
категории для всех возрастов ('E' и 'K-A') - '0+',  
категории от 3-х лет ('EC') - '3+',  
категории от 10-ти лет ('E10+') - '10+',  
категории от 13-ти лет ('Т') - '13+',  
категории от 17-ти лет ('М') - '17+',  
категории от 18-ти лет ('АО') - '18+',  
категории в состоянии "рейтинг ожидается" ('RP') - 'NaN'.  

```python
data['rating'] = data['rating'].str.replace('EC','3+')
data['rating'] = data['rating'].str.replace('E10','10')
data['rating'] = data['rating'].str.replace('K-A','0+')
data['rating'] = data['rating'].str.replace('T','13+')
data['rating'] = data['rating'].str.replace('M','17+')
data['rating'] = data['rating'].str.replace('AO','18+')
data['rating'] = data['rating'].str.replace('RP', 'NaN')
data['rating'] = data['rating'].str.replace('E','0+')
```

Проверим набор всех уникальных значений в столбце 'rating'.

```python
data['rating'].unique()
```

Подсчёт количества пропущенных значений в столбце 'rating'. 

```python
data['rating'].isna().sum()
```

Пропуски в столбце 'rating' заменяем на аббревиатуру 'no rating'.

```python
data['rating'] = data['rating'].fillna('no rating')
```

```python
data['rating'] = data['rating'].str.replace('NaN','no rating')
```

Пропуски в столбце 'rating' заменил на аббревиатуру 'no rating'.

Проверим набор всех уникальных значений в столбце 'rating' после изменений.

```python
data['rating'].unique()
```

Пропуски в столбце 'rating' заменяем на аббревиатуру 'no rating', так как удалять эти строки с пропусками нельзя - их довольно много, и заменять на среднее или медианное значение не возможно. Образовались эти пропуски возможно из-за того, что организация Entertainment Software Rating Board (ESRB) была основана только в 1994 году, а некоторые игры имеют более ранний период выхода на рынок, и рейтинга возрастной категории на момент выхода игры просто не существовало, либо игра была выпущена совсем недавно, и возрастной рейтинг еще не успели ей присвоить.

### Проверим гипотезы:
**1. Гипотеза.**

H0: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

H1: Средние пользовательские рейтинги платформ Xbox One и PC разные

```python
XOne = actual_data_general[(actual_data_general['platform'] == 'XOne') &  
                           (actual_data_general['user_score'] > 0)]['user_score']
PC = actual_data_general[(actual_data_general['platform'] == 'PC') & 
                         (actual_data_general['user_score'] > 0)]['user_score']
print('Средний пользовательский рейтинг платформы Xbox One:', XOne.mean())
print('Средний пользовательский рейтинг платформы PC:', PC.mean())
```

**Проверяем первую гипотезу**

```python
alpha = 0.05
results = st.ttest_ind(XOne, PC, equal_var=False)
print('p-значение:', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Вывод: 1 гипотеза имеет маловероятные результаты, поэтому эту гипотезу не используем. Средние пользовательские рейтинги платформ Xbox One и PC разные.**

**2. Гипотеза.**

H0: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») одинаковые

H1: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные

```python
Action = actual_data_general[(actual_data_general['genre'] == 'Action') & 
                             (actual_data_general['user_score'] > 0)]['user_score']
Sports = actual_data_general[(actual_data_general['genre'] == 'Sports') & 
                             (actual_data_general['user_score'] > 0)]['user_score']
print('Средний пользовательский рейтинг жанра Action:', Action.mean())
print('Средний пользовательский рейтинг жанра Sports:', Sports.mean())
```

**Проверяем вторую гипотезу**

```python
alpha = 0.05
results = st.ttest_ind(Action, Sports, equal_var=False)
print('p-значение:', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Вывод: Данные не противоречат 2 гипотезе. Средние пользовательские рейтинги жанров Action и Sports разные.**
